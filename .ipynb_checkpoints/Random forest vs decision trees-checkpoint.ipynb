### Random forest vs decision trees

Se usará el dataset *abalone.data*.

```python
import pandas as pd
import numpy as np
import graphviz #Para ver el arbol
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
from sklearn.metrics import precision_recall_fscore_support
from sklearn.ensemble import RandomForestClassifier
import seaborn as sn
%matplotlib inline
```

```python
df = pd.read_csv("abalone.data",header=None, na_values = "?")
df = df[df.columns[1:]] # quito la primera columna que no me interesa
df.head()
```

```python
# train y test
X_train, X_test, Y_train, Y_test = train_test_split(df[df.columns[0:-1]],df[df.columns[-1]], train_size=0.75)
```

```python
# pasarlo a matrices
X_trainM = X_train.as_matrix()
Y_trainM = Y_train.as_matrix()
X_testM = X_test.as_matrix()
Y_testM = Y_test.as_matrix()
```

### Decision tree

```python
# clasificador
clf = DecisionTreeClassifier()
clf = clf.fit(X_trainM, Y_trainM) # fit
# predict
Y_test_prediction = clf.predict(X_testM)
```

```python
conf = confusion_matrix(Y_testM, Y_test_prediction)
print("Matriz de confusión:")
df_cm = pd.DataFrame(conf)
sn.set()
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True)
plt.show()
```

```python
prec_rec = precision_recall_fscore_support(Y_testM, Y_test_prediction, average=None)

print("Evaluación para todas las clases:\nPrecision:\n{0} \nRecall:\n{1}".format(prec_rec[0],prec_rec[1]))
print("\nPromedio de evaluaciones:\nPrecision\n{0} \nRecall:\n{1}".format(np.mean(prec_rec[0]),np.mean(prec_rec[1])))
```

#### Vemos que un árbol de decisiones tiene en general mal desempeño con esta BD.

### Random forest

```python
clf2 = RandomForestClassifier()
clf2.fit(X_trainM, Y_trainM)
Y_test_prediction_RF = clf2.predict(X_testM)
```

```python
conf_RF = confusion_matrix(Y_testM, Y_test_prediction_RF)
print("Matriz de confusión:")

df_cm2 = pd.DataFrame(conf_RF)
sn.set()
plt.figure(figsize = (10,7))
sn.heatmap(df_cm2, annot=True)
plt.show()
```

```python
prec_rec_RF = precision_recall_fscore_support(Y_testM, Y_test_prediction_RF,average=None)
print("Evaluación para todas las clases:\nPrecision:\n{0} \nRecall:\n{1}".format(prec_rec_RF[0],prec_rec_RF[1]))
print("\nPromedio de evaluaciones:\nPrecision\n{0} \nRecall:\n{1}".format(np.mean(prec_rec_RF[0]),np.mean(prec_rec_RF[1])))
```

#### Observamos que un random forest tiene una pequeña mejora en comparación con un árbol de decisión.
### Si vemos ambas matrices de confusión observamos que los errores de clasificación son cercanos a la diagonal. Esto nos puede hacer pensar que si juntamos las clasificaciones, digamos de 2 en 2, los resultados puedan mejorar. Veámoslo...

```python
df2 = pd.read_csv("abalone.data",header=None, na_values = "?")
df2 = df2[df2.columns[1:]] # quito la primera columna que no me interesa

# para agrupar categorías (son los años de vida de los abulones)
categoria =[df2.iloc[i,7]//2 for i in range(df2.shape[0])] # categorías más manejables (agrupadas de dos en dos)
df2[9] = pd.Series(categoria)
df2.head()
```

### La última columna es la nueva clasificación a predecir

```python
# train y test
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(df2[df2.columns[0:-1]],df2[df2.columns[-1]], train_size=0.75)
# pasarlo a matrices
X_trainM2 = X_train2.as_matrix()
Y_trainM2 = Y_train2.as_matrix()
X_testM2 = X_test2.as_matrix()
Y_testM2 = Y_test2.as_matrix()
```

### Decision tree

```python
# clasificador
clf3 = DecisionTreeClassifier()
clf3 = clf3.fit(X_trainM2, Y_trainM2) # fit
```

```python
# predict
Y_test_prediction2 = clf3.predict(X_testM2)
conf3 = confusion_matrix(Y_testM2, Y_test_prediction2)
print("Matriz de confusión:")
df_cm3 = pd.DataFrame(conf3)
sn.set()
plt.figure(figsize = (14,9))
sn.heatmap(df_cm3, annot=True)
plt.show()
```

```python
prec_rec2 = precision_recall_fscore_support(Y_testM2, Y_test_prediction2, average=None)

print("Evaluación para todas las clases:\nPrecision:\n{0} \nRecall:\n{1}".format(prec_rec2[0],prec_rec2[1]))
print("\nPromedio de evaluaciones:\nPrecision\n{0} \nRecall:\n{1}".format(np.mean(prec_rec2[0]),np.mean(prec_rec2[1])))
```

#### Efectivamente el desempeño sube muchísimo si se juntan las clases de dos en dos.

### Random forest

```python
clf4 = RandomForestClassifier()
clf4.fit(X_trainM2, Y_trainM2)
Y_test_prediction_RF2 = clf4.predict(X_testM2)
```

```python
conf_RF2 = confusion_matrix(Y_testM2, Y_test_prediction_RF2)
print("Matriz de confusión:")

df_cm4 = pd.DataFrame(conf_RF2)
sn.set()
plt.figure(figsize = (14,9))
sn.heatmap(df_cm4, annot=True)
plt.show()
```

```python
prec_rec_RF2 = precision_recall_fscore_support(Y_testM2, Y_test_prediction_RF2,average=None)
print("Evaluación para todas las clases:\nPrecision:\n{0} \nRecall:\n{1}".format(prec_rec_RF2[0],prec_rec_RF2[1]))
print("\nPromedio de evaluaciones:\nPrecision\n{0} \nRecall:\n{1}".format(np.mean(prec_rec_RF2[0]),np.mean(prec_rec_RF2[1])))
```

#### Efectivamente el desempeño sube mucho si se juntan las clases de dos en dos. Además vemos que el desempeño mejora más con un árbol de decisión que con un bosque aleatorio.
